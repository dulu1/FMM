```python
import numpy as np
import matplotlib as mpl
import pylab as pl
import pickle
import pyekfmm as fmm
from scipy import interpolate
import matplotlib as mpl
from tqdm import tqdm
import pickle
import matplotlib.pyplot as plt

font_path = r'./Times New Roman.ttf'   
mpl.font_manager.fontManager.addfont(font_path)
mpl.rcParams['font.family'] = 'Times New Roman'
from scipy.interpolate import make_interp_spline

# Existing data
x = np.array([0.02,0.05,0.08,0.1,0.15,0.3])
y = np.array([5.62,2.06,1.39,1.52,1.87,6.52])

# Create a smooth curve
x_smooth = np.linspace(x.min(), x.max(), 500)
spl = make_interp_spline(x, y, k=3)
y_smooth = spl(x_smooth)

# Plot the smooth curve
plt.plot(x_smooth, y_smooth, label='Smooth Curve', color='b')
plt.scatter(x, y, color='r', label='Data Points')
plt.text(0.08, 1.2, f'({0.08},{1.39})')
plt.xlabel('relaxation factor',fontsize=14)
plt.ylabel('RMS/ms',fontsize=14)
plt.xticks(fontsize=14)  # 设置 x 轴刻度大小
plt.yticks(fontsize=14) 
#plt.title('Smooth Curve Plot')
#plt.legend()
plt.savefig('理论测试松弛系数图.png', bbox_inches='tight',dpi=800)
plt.show()
```
