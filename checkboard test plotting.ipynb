```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含随机数据的二维数组
data=np.random.uniform(-1, 1, (10, 10))

# 绘制网格
fig, ax = plt.subplots()
for i in range(10):
    for j in range(10):
        radius = abs(data[i, j]) * 0.4
        if data[i, j] > 0:
            circle = plt.Circle((j, i), radius, color='black')
        else:
            circle = plt.Circle((j, i), radius, color='white', fill=True, linestyle='-', linewidth=1)
        ax.add_artist(circle)

ax.set_facecolor('green')
# 设置坐标轴范围和刻度
ax.set_xlim(-1, 10)
ax.set_ylim(-1, 10)
ax.set_aspect('equal', adjustable='box')
ax.set_xticks(np.arange(0, 10, 1))
ax.set_yticks(np.arange(0, 10, 1))
ax.grid(False)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 图例中圆圈大小对应的数值
sizes = [380, 300, 220, 140, 140, 220,300,380]

# 圆圈大小对应的标签
labels = ['-5', '-4', '-3', '-2', '2', '3','4','5']

# 创建一个图和一组子图
fig, ax = plt.subplots()

# 绘制每个大小的圆圈，根据数值设置颜色和边缘颜色
for i, size in enumerate(sizes):
    color = 'black' if int(labels[i]) > 0 else 'white'
    edgecolor = 'black' if color == 'white' else None
    ax.scatter([], [], s=size, color=color, edgecolor=edgecolor, label=str(size))

# 添加图例，设置标签和标题
legend = ax.legend(title='dVp (%)', labels=labels, loc='center', bbox_to_anchor=(0.5, -0.1), ncol=len(sizes),title_fontsize='x-large')
for text in legend.get_texts():
    text.set_fontsize(14)  # 设置图例标签字体大小为14

plt.axis('off')  # 隐藏坐标轴
plt.savefig('图例.png', bbox_inches='tight',dpi=800)
plt.show()
```
