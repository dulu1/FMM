```python
import numpy as np
from scipy.interpolate import Rbf
import pyekfmm as fmm
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
from scipy import interpolate
import csv
import pickle

font_path = r'./Times New Roman.ttf'   
mpl.font_manager.fontManager.addfont(font_path)
mpl.rcParams['font.family'] = 'Times New Roman'
a=np.load('3.20测试算线中.npy')
b= np.load('3.20测试算线前.npy')
x=a-b
y1=x/b

a=np.load('3.20测试算线后.npy')
b= np.load('3.20测试算线中.npy')
x=a-b
y2=x/b
pp=5
m=y1[:,:,pp]
mm=y2[:,:,pp]

shape = (78, 36)
original_array = m
x, y = np.indices(shape)
coords = np.array([x[original_array != 0], y[original_array != 0]])
coords=(coords+1)*4
values = original_array[original_array != 0]
interp_func = Rbf(*coords, values, function='multiquadric', smooth=0.7)
shape=(312,144)
x_new, y_new = np.indices(shape)
all_coords = np.array([x_new.ravel(), y_new.ravel()])
interpolated_values = interp_func(*all_coords)
interpolated_array1 = interpolated_values.reshape(shape)
m=interpolated_array1[48:264, 32:104]

shape = (78, 36)
original_array = mm
x, y = np.indices(shape)
coords = np.array([x[original_array != 0], y[original_array != 0]])
coords=(coords+1)*4
values = original_array[original_array != 0]
interp_func = Rbf(*coords, values, function='multiquadric', smooth=0.7)
shape=(312,144)
x_new, y_new = np.indices(shape)
all_coords = np.array([x_new.ravel(), y_new.ravel()])
interpolated_values = interp_func(*all_coords)
interpolated_array1 = interpolated_values.reshape(shape)
mm=interpolated_array1[48:264, 32:104]

array1 = m
array2 = mm

result = np.zeros(array1.shape)
data=np.rot90(result)
for i in range(array1.shape[0]):
    for j in range(array1.shape[1]):
        if array1[i, j] > 0 and array2[i, j] > 0:
            result[i, j] = 2
        elif array1[i, j] < 0 and array2[i, j] < 0:
            result[i, j] = -2
        elif array1[i, j] > 0 and array2[i, j] < 0:
            result[i, j] = 1
        elif array1[i, j] < 0 and array2[i, j] > 0:
            result[i, j] = -1
        else:
            result[i, j]=0
colors = {2: 'orange', 1: 'peachpuff', 0: 'yellow', -1: 'lightcoral', -2: 'red'}

# 创建颜色数组
color_data = np.vectorize(colors.get)(data)

# 将颜色数组转换为数值类型
color_data_numeric = np.zeros_like(data, dtype=float)
for key, value in colors.items():
    color_data_numeric[color_data == value] = key

# 创建自定义颜色映射
cmap, norm = mcolors.from_levels_and_colors([-2, -1, 0, 1, 2, 3], ['red', 'orange', 'yellow', 'lightblue', 'blue'])

# 绘制图像
plt.imshow(color_data_numeric, cmap=cmap, norm=norm, interpolation='nearest')
plt.title('Color Coded 2D Array')
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), orientation='horizontal')
cbar.set_label('Color Legend')
plt.show()
print(result.shape)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
a_mid = np.load('3.20测试算线中.npy')
a_pre = np.load('3.20测试算线前.npy')
a_post = np.load('3.20测试算线后.npy')

# Calculate the differences and ratios
diff1 = (a_mid - a_pre) / a_pre
diff2 = (a_post - a_mid) / a_mid

# Extract values for comparison from a specific plane (e.g., plane at index 5)
array1 = diff1[:,:,4]
array2 = diff2[:,:,4]

# Define colors based on comparison results
colors = []
for x, y in zip(array1.flatten(), array2.flatten()):
    if x < 0 and y < 0:
        colors.append('orange')     # Both negative
    elif x > 0 and y > 0:
        colors.append('blue')       # Both positive
    elif x > 0 and y < 0:
        colors.append('lightblue')  # First positive, second negative
    elif x < 0 and y > 0:
        colors.append('peachpuff')  # First negative, second positive
    else:
        colors.append('yellow')     # Any other combination

# Create a 2D grid to plot the comparison
X, Y = np.meshgrid(range(array1.shape[0]), range(array1.shape[1]))

# Plot the comparison on a 2D grid
fig, ax = plt.subplots()
for i, (x, y, color) in enumerate(zip(X.flatten(), Y.flatten(), colors)):
    ax.add_patch(plt.Rectangle((x - 0.5, y - 0.5), 1, 1, color=color))

# Set the plot limits and labels
ax.set_xlim(-0.5, array1.shape[0] - 0.5)
ax.set_ylim(-0.5, array1.shape[1] - 0.5)
ax.set_xticks(range(array1.shape[0]))
ax.set_yticks(range(array1.shape[1]))
ax.set_title('Plane Comparison in a 2D Grid')

# Display the plot
plt.show()
```
